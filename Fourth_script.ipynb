```
import datetime
from datetime import datetime
import pandas as pd
import requests
from pandas import DataFrame
from pandas import ExcelWriter

#Traigo el CVS de internet

def wget(url):
    r = requests.get(url, allow_redirects=True)
    with open(url[url.rfind('/') + 1::], 'wb') as f:
        f.write(r.content)

wget ("https://covid.ourworldindata.org/data/ecdc/full_data.csv")

df1 = pd.read_csv('full_data.csv')
df2 = DataFrame(df1, columns=['date','location','total_cases','total_deaths'])

#Intervalo de tiempo

df_tiempo = pd.read_csv('full_data.csv', header=0)
datos_records = df_tiempo.to_dict('records')

t1 = []
setDates = set()

for record in datos_records:
    if record['location'] == 'World':
        t1.append(record)

s1 = dict()

for caso in t1:
    s1[caso['date']] = caso['total_cases']

for x in s1:
    setDates.add(x)

listDates = list(setDates)
listDates.sort(key=lambda date: datetime.strptime(date, '%Y-%m-%d'))
listDates_acotada = list()
listDates_acotada.sort(key=lambda date: datetime.strptime(date, '%Y-%m-%d'))

print("Este programa hace una lista por día de mayor a menor de casos de COVID-19 por país en un intervalo de tiempo (en Excel).")
print("Seleccione la fecha de comienzo del intervalo:")
year_ini = int(input('Ingrese el año: '))
month_ini = int(input('Ingrese el mes: '))
day_ini = int(input('Ingrese el dia: '))
print("Seleccione la fecha de fin del intervalo:")
year_end = int(input('Ingrese el año: '))
month_end = int(input('Ingrese el mes: '))
day_end = int(input('Ingrese el dia: '))
date_ini_dt = datetime(year_ini, month_ini, day_ini)
date_end_dt = datetime(year_end, month_end, day_end)

date_ini = date_ini_dt.strftime('%Y-%m-%d')
date_end = date_end_dt.strftime('%Y-%m-%d')

print("Fecha de inicio")
print(date_ini)
print("Fecha final")
print(date_end)

for i in range(1, len(listDates)):
    if (date_ini < listDates[i] and date_end > listDates[i]):
        listDates_acotada.append(listDates[i])

#Armo los excel con lo siguiente

dateprueba = list()
dateprueba = listDates_acotada
print("Se realizará un excel con las siguientes fechas:")
print(dateprueba)

list_location = []
list_cases = []
list_deaths = []

writer = ExcelWriter('lista_covid.xlsx')
print("Esto puede tardar un tiempo.")

for z in range(0, len(dateprueba)):
    dateprobar = 0
    dateprobar = dateprueba[z]
    for i in range(0, len(df2["date"])):
        if dateprobar == (df2["date"][i]):
            list_location.append(df2["location"][i])
            list_cases.append(df2["total_cases"][i])
            list_deaths.append(df2["total_deaths"][i])
            df_location = pd.DataFrame(list_location)
            df_cases = pd.DataFrame(list_cases)
            df_deaths = pd.DataFrame(list_deaths)
            df_concat = pd.concat([df_location, df_cases, df_deaths], axis=1, sort=False)
            df_concat.columns = ('Lugar', 'Casos totales', 'Fallecimientos')
            result = df_concat.sort_values(by=['Casos totales'], ascending=False)
    result.to_excel(writer, sheet_name='Diat ' + dateprueba[z])
    writer.save()
```
